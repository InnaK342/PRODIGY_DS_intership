Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.


Sample Dataset :- https://data.worldbank.org/indicator/SP.POP.TOTL

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

##Reading data from the WorldBank Data.xlsx file

```python
data = pd.read_excel('WorldBank Data.xlsx')
```

##Plotting a histogram

```python
sns.histplot(data['Life expectancy at birth (years)'])
plt.title('Histogram of the age distribution of all countries from 1991 to 2014')
plt.ylabel('Frequency');
```

```python
data_2014 = data.loc[(data['Year']==2014)]
data_2014.head()
```

```python
data_2014.shape
```

```python
sns.histplot(data=data_2014['Life expectancy at birth (years)'])
plt.title('Histogram of the age distribution of all countries in 2014')
plt.ylabel('Frequency')
plt.xlabel('Life expectancy at birth (years) in 2014');
```

##Plotting a barplot

```python
data_group = data.groupby('Country Name')['Life expectancy at birth (years)'].mean()
data_sort = data_group.sort_values(ascending=False)[:10]
data_sort
```

```python
sns.barplot(data=data_sort, color='skyblue')
plt.title('10 countries with the highest average life expectancy from 1991 to 2014')
plt.xlabel('Country Name')
plt.ylabel('Life expectancy')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(75, 85))
plt.ylim(75, 85)
plt.show()
```

```python
data_group_region = data.groupby('Region')['Life expectancy at birth (years)'].mean()
data_sort_region = data_group_region.sort_values(ascending=False)
data_sort_region
```

```python
sns.barplot(data=data_sort_region, color='skyblue')
plt.title('Average life expectancy by region')
plt.xlabel('Region')
plt.ylabel('Life expectancy')
plt.xticks(rotation=45, ha='right')
plt.yticks(np.arange(50, 84, 3))
plt.ylim(50, 84)
plt.show()
```

##Plotting a pieplot


```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:7]
plt.pie(data_sort_region, labels=data_sort_region.index, colors=colors, autopct='%.0f%%')
plt.title('Average life expectancy by region')
plt.axis('equal')
plt.tight_layout()
plt.show()
```
